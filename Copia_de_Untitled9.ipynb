```python
import numpy as np

def f(x):
    return 2*x**3 + 3*x**2 - 2*x + 1

def f_prime(x):
    return 6*x**2 + 6*x - 2
```

```python
def bisection_method(f, a, b, tol=1e-6, max_iter=100):
    iterations = 0
    while abs(b - a) > tol and iterations < max_iter:
        c = (a + b) / 2
        if f(c) == 0:
            break
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
        iterations += 1
    return (a + b) / 2, iterations
```

```python
def newton_raphson_method(f, f_prime, x0, tol=1e-6, max_iter=100):
    iterations = 0
    x = x0
    while abs(f(x)) > tol and iterations < max_iter:
        x = x - f(x) / f_prime(x)
        iterations += 1
    return x, iterations
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy
%matplotlib inline
def calcularGradiente(fU):
    d1fU = sympy.diff(fU)
    print("Función original: ",fU)
    print("Derivada de la función : ",d1fU)
    return None
```

```python
def fx(x):
  return (2*x**3)-(2*x)-((4*x**2)**0.5)

def d1Fx(x):
  return  6*x**2 + 2*x - 6
```

**MÉTODO DE BISECCIÓN**

```python
def biseccion(a,b,epsilon):

    cont = 0
    registro = []
    x_list, y_list = [a], [fx(a)]

    while(True):

        alpha = (a+b)/2

        # Primera derivada en a
        dU_a = d1Fx(a)

        #Primera derivada en alpha
        dU_alpha = d1Fx(alpha)

        if(dU_a*dU_alpha < 0):
            b = alpha
        else:
            a = alpha

        x_list.append(a)
        y_list.append(fx(a))

        # Condicion de finalizacion
        if(np.abs(a-b) < epsilon):
            print('-------------------------------------------------------------')
            print("It: {:02} - Temp: {:.10f} - Costo {:.10f}".format(cont, a, fx(a)))
            break

        cont = cont + 1
        registro.append([cont, a, fx(a)])
        print("It: {:02} - Temp: {:.10f} - Costo {:.10f}".format(cont, a, fx(a)))

    return x_list, y_list
```

```python
x_list, y_list = biseccion(0.5,1.5,0.001)
```

El minimo valor: 0.8466796875 * ----> ESTE ES EL MEJOR

```python
def graficaPuntos(x_list,y_list,fx,a,b):
    x = np.linspace(a,b,100)
    plt.plot()
    plt.scatter(x_list,y_list,c="g")
    plt.plot(x,fx(x), c="r")
    plt.title("La gráfica de los puntos encontrados sobre la función")
    plt.show()
```

```python
graficaPuntos(x_list,y_list,fx,0,1.5)
```

```python
def calcularGradientes(fU):
    d1fU = sympy.diff(fU)
    d2fU = sympy.diff(d1fU)
    print("Función original: ",fU)
    print("Primera derivada :",d1fU)
    print("Segunda derivada :",d2fU)
    return None
```

```python
x = sympy.Symbol('x')
fU = (2*x**3)-(2*x)-((4*x**2)**0.5)
calcularGradientes(fU)
```

```python
def d2Fx(x):
  return 12*x + 1.5
```

```python
def NewtonRaphson(a,b, epsilon):

    x = a
    cont = 0
    registro = []
    x_list,y_list=[a],[fx(a)]
    while True:

        d1U_x = d1Fx(x)
        d2U_x = d2Fx(x)

        xprev = x
        x = xprev - d1U_x/d2U_x

        cont = cont + 1
        registro.append([cont, x, fx(x)])
        x_list.append(x)
        y_list.append(fx(x))

        print("It: {:02d} - Temp: {:.10f} - Costo: {:.10f}".format(cont, x, fx(x)))

        if(np.abs(x - xprev) <= epsilon):
            print("----------------------------------------------------------------")
            print("It: {:02d} - Temp: {:.10f} - Costo: {:.10f}".format(cont, x, fx(x)))
            break

    return x_list,y_list
```

```python
x_list,y_list = NewtonRaphson(0.5,1.5, 0.001)
```

El minimo valor: 0.8471306401 * ----> ESTE ES EL MEJOR

**MÉTODO DEL DESCENSO DEL GRADIENTE**

```python
def gradient(x_start, precision, l_r):

    x_list, y_list = [x_start], [function(x_start)]

    # Se realiza un bucle While

    while True:

        d_x = - d1Fx(x_start)

        # Obtenemos el nuevo valor de x sumando el valor previo de x más la multiplicación del l_r por la derivada.
        x_start += (l_r * d_x)

        # Agregamos el nuevo valor de x a la lista para posteriormente visualizarlo.
        x_list.append(x_start)

        # Agregamos el nuevo valor de y a la lista para posteriormente visualizarlo.
        y_list.append(function(x_start))

        print("It: {:02d} - Temp: {:.10f} - Costo: {:.10f}".format(len(x_list), x_start, fx(x_start)))
        # Salimos del bucle cuando se cumpla la condición.
        if abs(x_list[-1] - x_list[-2]) <= precision:
          print("----------------------------------------------------------------")
          print("It: {:02d} - Temp: {:.10f} - Costo: {:.10f}".format(len(x_list), x_start, fx(x_start)))
          break

    return x_list, y_list
```

```python
x_list, y_list = gradient(0.5,1.5,0.01)
```

El minimo valor: 0.5350000000 * ----> ESTE ES EL MEJOR
